```python
import pandas as pd

df = pd.read_csv("../data/raw/booksummaries.csv", header=None, sep='|')

df = df[0].str.split(r'\t', expand=True)

df.columns = ["Wiki_ID", "Freebase_ID", "Title", "Author", "Pub_Date", "Genres", "Summary"]

df = df.drop(columns=['Wiki_ID', 'Freebase_ID', 'Pub_Date'])

df.head()
```

```python
df.isnull().sum()
```

```python
import ast

def extract_genres(genre_str):
    try:
        genre_dict = ast.literal_eval(genre_str)
        return list(genre_dict.values())
    except:
        return []

df['Genres'] = df['Genres'].apply(extract_genres)

df.head()
```
